```python
def pretty_bar(labels, results, title=''):
    plt.figure(dpi=200)
    plt.bar(labels, results, 0.4)
    plt.ylim(0, 25)

    for i, v in enumerate(labels):
        plt.gca().text(i - 0.1, 
                  results[i] - 5, 
                  results[i], 
                  fontsize=12,
                  color='w')

    plt.title(title)
    
    
labels = ['Trained Policy Model', 'MPC', 'MPC with GT future']
results = [20.0, 13.4, 6.8]
pretty_bar(labels, results, 'Comparison of error rates')
```
